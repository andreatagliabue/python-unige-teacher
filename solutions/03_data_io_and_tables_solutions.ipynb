## 6. From force to strength (calculated columns)

The machine only records **force** and **displacement**. But "strength" is **stress** (force spread over the cross-section), and stretch is **strain**. So we compute them from what we have:

```text
area_mm2   = π × (diameter_mm / 2)²
stress_MPa = force_N / area_mm2
strain     = displacement_mm / gauge_length_mm
```

Handy unit fact: **1 N/mm² = 1 MPa**, so force in newtons ÷ area in mm² gives stress directly in MPa.

In pandas we create a whole new column in one line — the calculation runs on every row at once.

````python
# Create calculated columns

df["area_mm2"] = math.pi * (df["diameter_mm"] / 2) ** 2
df["stress_MPa"] = df["force_N"] / df["area_mm2"]
df["strain"] = df["displacement_mm"] / df["gauge_length_mm"]

# Inspect the new columns
df[["sample_id", "force_N", "displacement_mm", "area_mm2", "stress_MPa", "strain"]].head()
````

````python
# Check ranges of the new quantities
print("Stress range [MPa]:")
print(df["stress_MPa"].min(), "to", df["stress_MPa"].max())
print()

print("Strain range [-]:")
print(df["strain"].min(), "to", df["strain"].max())
````

## 7. Find the glitches (data-quality checks)

Raw machine exports are rarely perfect — sensors hiccup and readings drop out. Before trusting any result, we look for obvious nonsense:

- **missing** values (`NaN`);
- **negative** forces or displacements (physically impossible here);
- unexpected sample names or treatment labels.

Let's see what this file is hiding.

````python
# Count missing values in each column
df.isna().sum()
````

````python
# Check unique samples and treatments
print("Sample IDs:")
print(df["sample_id"].unique())
print()

print("Treatments:")
print(df["treatment"].unique())
````

````python
# Check for negative values in quantities that should not be negative
negative_force_rows = df[df["force_N"] < 0]
negative_displacement_rows = df[df["displacement_mm"] < 0]

print("Rows with negative force:", len(negative_force_rows))
print("Rows with negative displacement:", len(negative_displacement_rows))
````

### From "raw" to "clean": removing bad rows

The checks above are not just for show — this file really does contain a few bad measurements: some **missing** force readings (`NaN`), a couple of **negative** forces, and one negative displacement (sensor glitches). None of these make physical sense.

We remove them **step by step**, keeping the original `df` and building a new table called `df_clean`. Each step is just a filter like the ones in section 5.

````python
# Start from the raw table and remove bad rows, one step at a time.

# 1) drop rows where the force reading is missing (NaN)
df_clean = df.dropna(subset=["force_N"])

# 2) keep only rows with a non-negative force
df_clean = df_clean[df_clean["force_N"] >= 0]

# 3) keep only rows with a non-negative displacement
df_clean = df_clean[df_clean["displacement_mm"] >= 0]

# How many rows did we drop?
print("Rows before:", len(df))
print("Rows after: ", len(df_clean))
print("Removed:    ", len(df) - len(df_clean))
````

## 8. One number per specimen — and the answer

We now have a clean table with ~120 measurement points per specimen. To compare specimens we need **one number each**: the **maximum stress** reached, i.e. the strength.

`groupby("sample_id")` splits the rows into one group per specimen. Then we pick the `stress_MPa` column and take its `max()` inside each group. We do it in two small steps.

```text
many rows per specimen → one number per specimen
```

````python
# Step 1: group by specimen and take the maximum stress in each group
max_stress_by_sample = df_clean.groupby("sample_id")["stress_MPa"].max()

# Step 2: sort from strongest to weakest
max_stress_by_sample = max_stress_by_sample.sort_values(ascending=False)

max_stress_by_sample
````

### So, which treatment wins?

Let's average the peak stress *within each treatment* and sort from strongest to weakest.

````python
# Same idea, but grouping by treatment instead of by specimen

# Step 1: highest stress reached within each treatment
max_stress_by_treatment = df_clean.groupby("treatment")["stress_MPa"].max()

# Step 2: sort from strongest to weakest
max_stress_by_treatment = max_stress_by_treatment.sort_values(ascending=False)

max_stress_by_treatment
````

> 🎯 **The answer.** The **quenched** specimen (S03) reaches the highest stress (~516 MPa), while the **annealed** ones are the softest (~355–367 MPa). So, for this dataset, **quenching gives the strongest material**.
>
> We started from a raw, slightly broken machine export and turned it into a one-line conclusion — that is exactly what this whole workflow is for.

### (Optional) Several summaries at once with `.agg()`

The two steps above already answered our question. When you need **several** numbers per specimen at the same time (treatment, max stress, max strain, …), pandas can do it in one go with `.agg()`. It is handy but a bit more advanced — feel free to skip it for now and come back later.

````python
# Optional / advanced: one row per specimen with several columns at once
summary_by_sample = (
    df_clean
    .groupby("sample_id")
    .agg(
        treatment=("treatment", "first"),
        max_stress_MPa=("stress_MPa", "max"),
        max_strain=("strain", "max"),
    )
    .reset_index()
    .sort_values("max_stress_MPa", ascending=False)
)

summary_by_sample
````

## Optional extension — Hardness data

The folder also contains a smaller dataset:

```text
data/hardness_positions.csv
```

This dataset contains repeated hardness measurements at several positions on the same samples.

Columns:

- `sample_id`;
- `treatment`;
- `position_id`;
- `hardness_HV`;
- `is_valid` (some measurements were flagged as invalid).

Possible tasks:

1. read the file;
2. keep only rows where `is_valid == True`;
3. calculate mean and standard deviation of hardness for each sample;
4. calculate mean hardness for each treatment;
5. save a summary CSV.

````python
# Optional extension — SOLUTION

hardness = load_table("hardness_positions.csv")
valid_hardness = hardness[hardness["is_valid"]]

# Mean and std per sample
stats_by_sample = (
    valid_hardness.groupby("sample_id")["hardness_HV"]
    .agg(mean="mean", std="std")
    .reset_index()
)
print("Hardness by sample:")
print(stats_by_sample.to_string(index=False))

# Mean per treatment
mean_by_treatment = (
    valid_hardness.groupby("treatment")["hardness_HV"]
    .mean()
    .reset_index()
    .rename(columns={"hardness_HV": "mean_HV"})
)
print()
print("Mean hardness by treatment:")
print(mean_by_treatment.to_string(index=False))

# Save
from pathlib import Path
results_dir = Path("results")
results_dir.mkdir(exist_ok=True)
stats_by_sample.to_csv(results_dir / "hardness_by_sample.csv", index=False)
print()
print("Saved to results/hardness_by_sample.csv")
````
