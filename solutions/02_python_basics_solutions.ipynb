# 2. Arithmetic operators

Python can be used as a calculator, but the real advantage is that calculations can be saved, repeated, checked, and modified.

Common arithmetic operators:

| Operator | Meaning | Example with `a = 7`, `b = 2` | Result |
|---|---|---:|---:|
| `+` | addition | `a + b` | `9` |
| `-` | subtraction | `a - b` | `5` |
| `*` | multiplication | `a * b` | `14` |
| `/` | division | `a / b` | `3.5` |
| `//` | integer division | `a // b` | `3` |
| `**` | power | `a ** b` | `49` |
| `%` | remainder | `a % b` | `1` |

For scientific calculations, `**` is especially useful for powers.

````python
a = 7
b = 2

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b
integer_division = a // b
power = a ** b
remainder = a % b

print("a =", a)
print("b =", b)
print()

print("a + b  =", addition)
print("a - b  =", subtraction)
print("a * b  =", multiplication)
print("a / b  =", division)
print("a // b =", integer_division)
print("a ** b =", power)
print("a % b  =", remainder)
````

## Example: circular area

If a tensile specimen has a circular cross-section, the area is:

```text
area = π × radius²
```

where:

```text
radius = diameter / 2
```

````python
diameter_mm = 5.0

radius_mm = diameter_mm / 2
area_mm2 = math.pi * radius_mm ** 2

print("Diameter:", diameter_mm, "mm")
print("Radius:", radius_mm, "mm")
print("Area:", area_mm2, "mm^2")
````

## Making numbers readable: `round()` and f-strings

Did you notice the area printed as `19.634954084936208`? That is far more digits than we need. Two simple tools make scientific output readable:

- `round(value, n)` returns the value rounded to `n` decimals;
- an **f-string** lets us build a piece of text with values inserted into it: put an `f` before the quotes and write the value inside `{ }`. Adding `:.1f` fixes the number of decimals (here, 1).

````python
# round() returns a rounded number
print(round(area_mm2, 2))

# f-string: insert values directly into the text, with a chosen number of decimals
print(f"Area = {area_mm2:.2f} mm^2")
print(f"Area = {area_mm2:.1f} mm^2   (only one decimal)")

# values and calculations can go inside the { }
print(f"Radius = {radius_mm} mm   ->   area = {math.pi * radius_mm**2:.1f} mm^2")
````

## Exercise 2.1 — Calculate stress and strain

A tensile specimen is pulled in the machine. For one measurement point you have:

- force: 1500 N
- diameter: 5.0 mm
- displacement: 0.25 mm
- gauge length: 25.0 mm

Compute, step by step:

1. the cross-sectional **area** (circular section — same formula as the example above);
2. the **stress**: the force divided by the area;
3. the **strain**: the displacement divided by the gauge length.

Then **print the stress and the strain with two decimal places**
(use `round(value, 2)` or an f-string with `:.2f`).

Note: because `1 N/mm² = 1 MPa`, the stress comes out directly in MPa.

````python
# Exercise 2.1 — SOLUTION

force_N = 1500
diameter_mm = 5.0
displacement_mm = 0.25
gauge_length_mm = 25.0

area_mm2 = math.pi * (diameter_mm / 2) ** 2
stress_MPa = force_N / area_mm2
strain = displacement_mm / gauge_length_mm

print(f"Stress:  {stress_MPa:.2f} MPa")
print(f"Strain:  {strain:.2f}")
````

## Boolean logic

We can combine comparisons using:

- `and`;
- `or`;
- `not`.

````python
stress_MPa = 450
strain = 0.015

high_strength = stress_MPa > 400
low_strain = strain < 0.02

print("High strength:", high_strength)
print("Low strain:", low_strain)
print()

print("High strength AND low strain:", high_strength and low_strain)
print("High strength OR low strain:", high_strength or low_strain)
print("NOT high strength:", not high_strength)
````

# 6. Lists

A list stores multiple values.

Lists are useful for repeated measurements, sample names, temperatures, file names, and many other cases.

We reach a single element by its **position** (the *index*). In Python counting starts at `0`:

- `values[0]` is the first element;
- `values[1]` is the second;
- `values[-1]` is the **last** element (negative indices count from the end);
- `values[-2]` is the second to last.

````python
hardness_values = [220, 225, 223, 221]

print("All values:", hardness_values)
print("First value:", hardness_values[0])
print("Second value:", hardness_values[1])
print("Last value:", hardness_values[-1])
print("Number of values:", len(hardness_values))
print("Minimum:", min(hardness_values))
print("Maximum:", max(hardness_values))

mean_hardness = sum(hardness_values) / len(hardness_values)
print(f"Mean: {mean_hardness:.1f}")
````

# 8. Functions

A function is a reusable block of code with a name.

In scientific work, functions are especially useful for formulas.

A function can have:

- inputs;
- calculations;
- an output.

````python
def compute_stress(force_N, area_mm2):
    stress_MPa = force_N / area_mm2
    return stress_MPa

stress = compute_stress(1200, 20)

print("Stress:", stress, "MPa")
````

## Example: functions for a tensile test

Let us define three small functions:

- circular area;
- stress;
- strain.

It is good practice to begin a function with a short **docstring**: a line of text between triple quotes `"""..."""`, placed right under the `def` line, that says what the function does. It is optional, but it documents the function for whoever reads it later (including you in a month). Typing `help(function_name)` shows it.

````python
def compute_circular_area(diameter_mm):
    """Return the area (mm^2) of a circular cross-section from its diameter (mm)."""
    radius_mm = diameter_mm / 2
    area_mm2 = math.pi * radius_mm ** 2
    return area_mm2

def compute_stress(force_N, area_mm2):
    """Return the stress (MPa) given force (N) and area (mm^2)."""
    return force_N / area_mm2

def compute_strain(displacement_mm, gauge_length_mm):
    """Return the strain (dimensionless) given displacement and gauge length (mm)."""
    return displacement_mm / gauge_length_mm

area = compute_circular_area(5.0)
stress = compute_stress(1500, area)
strain = compute_strain(0.25, 25.0)

print(f"Area = {area:.2f} mm^2")
print(f"Stress = {stress:.1f} MPa")
print(f"Strain = {strain:.4f}")

# help() shows the function's docstring
help(compute_circular_area)
````

## From raw data to stress-strain data

We can **reuse the functions we defined in section 8**
(`compute_circular_area`, `compute_stress`, `compute_strain`).
If you get a `NameError` here, run the cells in section 8 first.

To go through the two lists `force_N` and `displacement_mm` *together* —
each force with its matching displacement — we use `zip()`:

```python
for force, displacement in zip(force_N, displacement_mm):
    ...
```

`zip()` pairs up the items: the 1st force with the 1st displacement,
the 2nd with the 2nd, and so on.

````python
# Requires the functions defined in section 8
diameter_mm = 5.02          # sample S01
gauge_length_mm = 25.0

area_mm2 = compute_circular_area(diameter_mm)

stress_MPa = []
strain = []

# zip() lets us loop over the two lists in parallel (each force with its displacement)
for force, displacement in zip(force_N, displacement_mm):
    stress_value = compute_stress(force, area_mm2)
    strain_value = compute_strain(displacement, gauge_length_mm)
    stress_MPa.append(stress_value)
    strain.append(strain_value)
    # f-string keeps the table readable (4 and 1 decimals)
    print(f"strain = {strain_value:.4f}   stress = {stress_value:.1f} MPa")
````

````python
plt.figure(figsize=(7, 4))
plt.plot(strain, stress_MPa, marker="o", label="Sample S01")

plt.xlabel("Strain [-]")
plt.ylabel("Stress [MPa]")
plt.title("Stress-strain curve")
plt.grid(True)
plt.legend()

plt.show()
````

## Optional Exercise 12.1

Write a function called `compute_stress_strain_curve`.

Inputs:

- `force_values_N`;
- `displacement_values_mm`;
- `diameter_mm`;
- `gauge_length_mm`.

Output:

- `strain_values`;
- `stress_values_MPa`.

Then use it for both samples and plot both curves on the same figure.

````python
# Optional Exercise 12.1 — SOLUTION

def compute_stress_strain_curve(force_values_N, displacement_values_mm,
                                 diameter_mm, gauge_length_mm):
    '''Return (strain_values, stress_values_MPa) for a tensile test.'''
    area_mm2 = math.pi * (diameter_mm / 2) ** 2
    strain_values     = [d / gauge_length_mm for d in displacement_values_mm]
    stress_values_MPa = [f / area_mm2        for f in force_values_N]
    return strain_values, stress_values_MPa

strain_1, stress_1 = compute_stress_strain_curve(
    sample_1_force_N, sample_1_displacement_mm,
    sample_1_diameter_mm, sample_1_gauge_length_mm,
)
strain_2, stress_2 = compute_stress_strain_curve(
    sample_2_force_N, sample_2_displacement_mm,
    sample_2_diameter_mm, sample_2_gauge_length_mm,
)

plt.figure()
plt.plot(strain_1, stress_1, label=f"{sample_1_id} ({sample_1_treatment})")
plt.plot(strain_2, stress_2, label=f"{sample_2_id} ({sample_2_treatment})")
plt.xlabel("Strain [-]")
plt.ylabel("Stress [MPa]")
plt.title("Stress-strain curves: S01 vs S03")
plt.legend()
plt.grid(True)
plt.show()
````
